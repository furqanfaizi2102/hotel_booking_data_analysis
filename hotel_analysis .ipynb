# DATA ANALYSIS AND VISUALIZATION

```python
cancelled_per = df['is_canceled'].value_counts(normalize = True)
print(cancelled_per)
```

```python
plt.figure(figsize = (5,4))
plt.bar(['Cancelled','Not Cancelled'],df['is_canceled'].value_counts())
plt.title("Reservation status count")
plt.show()
```

```python
sns.countplot(data = df,x='hotel',hue = 'is_canceled', palette='rocket')
plt.ylabel("Reservation Count")
plt.title('Reservation Status in Different Hotel')
plt.xlabel('Hotel')
plt.show()
```

```python
city_hotel = df[df['hotel'] == 'City Hotel']
city_hotel_per=city_hotel['is_canceled'].value_counts(normalize = True)

resort_hotel = df[df['hotel'] == 'Resort Hotel']
resort_hotel_per = resort_hotel['is_canceled'].value_counts(normalize = True)

print(city_hotel_per)
print(resort_hotel_per)
```

```python
plt.subplot(1,2,1)
palette_color = sns.color_palette('rocket')
plt.pie(x=city_hotel_per,labels = ['Not Canceled','Canceled'],autopct='%.0f%%' )
plt.title('City Hotel Cancellation')
plt.subplot(1,2,2)
palette_color = sns.color_palette('rocket')
plt.pie(x=resort_hotel_per,labels = ['Not Canceled','Canceled'],autopct='%.0f%%' )
plt.title('Resort Hotel Cancellation')
plt.show()
```

```python
resort_hotel_adr = resort_hotel.groupby('reservation_status_date')[['adr']].mean()
city_hotel_adr = city_hotel.groupby('reservation_status_date')[['adr']].mean()
```

```python
plt.figure(figsize=(20,10))
plt.title('Daily Average Rate in City and Resort Hotel')
plt.plot(resort_hotel_adr.index,resort_hotel_adr,label = 'Resort Hotel')
plt.plot(city_hotel_adr.index,city_hotel_adr,label = 'City Hotel')
plt.legend(fontsize = 20)
plt.show()
```

```python
df['month'] = df['reservation_status_date'].dt.month

plt.figure(figsize = (16,8))
sns.countplot(x= 'month',data = df,hue = 'is_canceled',palette = 'rocket')
plt.title("Reservation Per month")
plt.xlabel('Month')
plt.ylabel('Reservation')
plt.legend(["Not canceled","Canceled"])
plt.show()
```

```python
plt.figure(figsize=(16,8))
sns.barplot(x= 'month',y='adr', data = df[df['is_canceled'] == 1].groupby('month')[['adr']].sum().reset_index(), palette='rocket')
plt.title('ADR per Month')
plt.show()
```

```python
cancelled_data = df[df['is_canceled'] == 1]
not_cancelled_data = df[df['is_canceled'] == 0]

top_10_canceled_country = df['country'].value_counts(ascending = False)[:10]
```

```python
plt.figure(figsize =(7,7))
plt.title('Top 10 countries with reservation cancelation')
plt.pie(top_10_canceled_country,autopct ='%.2f',labels = top_10_canceled_country.index)
plt.show()
```

```python
df['market_segment'].value_counts(normalize = True)
```

```python
cancelled_data['market_segment'].value_counts(normalize = True)
```

```python
cancelled_data_adr = cancelled_data.groupby('reservation_status_date')[['adr']].mean()
cancelled_data_adr.reset_index(inplace = True)
cancelled_data_adr.sort_values('reservation_status_date',inplace = True)

not_cancelled_data_adr = not_cancelled_data.groupby('reservation_status_date')[['adr']].mean()
not_cancelled_data_adr.reset_index(inplace = True)
not_cancelled_data_adr.sort_values('reservation_status_date',inplace = True)

plt.figure(figsize = (20,6))
plt.title('Daily Average ADR',fontsize = 20)
plt.plot(cancelled_data_adr['reservation_status_date'],cancelled_data_adr['adr'],label = "Cancelled")
plt.plot(not_cancelled_data_adr['reservation_status_date'],not_cancelled_data_adr['adr'],label = "Not Cancelled")
plt.legend()
plt.show()
```
